```python
from abc import ABC, abstractmethod

class BuildingSystem(ABC):
    @abstractmethod
    def start(self):
        pass

    @abstractmethod
    def stop(self):
        pass

    @abstractmethod
    def status(self):
        pass

class AirConditioningSystem(BuildingSystem):
    def start(self):
        print("AC started")

    def stop(self):
        print("AC stopped")

    def status(self):
        print("AC status: running")

class LightingSystem(BuildingSystem):
    def start(self):
        print("Lights on")

    def stop(self):
        print("Lights off")

    def status(self):
        print("Lighting status: active")

class SecuritySystem(BuildingSystem):
    def start(self):
        print("Security armed")

    def stop(self):
        print("Security disarmed")

    def status(self):
        print("Security status: secure")

class FireAlarmSystem(BuildingSystem):
    def start(self):
        print("Fire alarm activated")

    def stop(self):
        print("Fire alarm deactivated")

    def status(self):
        print("Fire alarm status: normal")

systems = [
    AirConditioningSystem(),
    LightingSystem(),
    SecuritySystem(),
    FireAlarmSystem()
]

for s in systems:
    s.start()
    s.status()
    s.stop()
```
